```python
from procyclingstats import Stage

missing_race_names = {
    "2017 Paris-Roubaix": "race/paris-roubaix/2017/result",
    "2018 Paris-Roubaix": "race/paris-roubaix/2018/result",
    "2019 Paris-Roubaix": "race/paris-roubaix/2019/result",
    "2021 Paris-Roubaix": "race/paris-roubaix/2021/result",
    "2022 Paris-Roubaix": "race/paris-roubaix/2022/result",
    "2023 Paris-Roubaix": "race/paris-roubaix/2023/result",

    "2017 Amstel Gold Race": "race/amstel-gold-race/2017/result",
    "2018 Amstel Gold Race": "race/amstel-gold-race/2018/result",
    "2019 Amstel Gold Race": "race/amstel-gold-race/2019/result",
    "2021 Amstel Gold Race": "race/amstel-gold-race/2021/result",
    "2022 Amstel Gold Race": "race/amstel-gold-race/2022/result",
    "2023 Amstel Gold Race": "race/amstel-gold-race/2023/result"}

data = []

for race_name, race_url in missing_race_names.items():
    try:
        stage = Stage(race_url)
        won_how = stage.won_how() 
        data.append({"Race": race_name, "Won How": won_how})
    except Exception as e:
        print(f"Error fetching {race_name}: {e}")
        data.append({"Race": race_name, "Won How": "Error fetching data"})

df = pd.DataFrame(data)
df.to_excel("pcs_missing_race_how_data.xlsx", index=False)
```

```python
from procyclingstats import Stage
import pandas as pd
import re

missing_race_names
data = []

for race_name, race_url in missing_race_names.items():
    try:
        stage = Stage(race_url)

        avg_speed = stage.avg_speed_winner()
        results = stage.results("rank", "rider_name", "team_name", "time")

        while len(results) < 5: 
            results.append({"rider_name": "N/A", "team_name": "N/A", "time": "N/A"})

        top_5 = results[:5]

        winner_time = top_5[0]["time"]

        time_gaps = [f"{rider['rider_name']} - {rider['time']}" for rider in top_5[1:]]

        data.append({
            "Race": race_name,
            "Winner": top_5[0]["rider_name"],
            "Team": top_5[0]["team_name"],
            "Winner Time": winner_time,
            "Avg Speed (km/h)": avg_speed if avg_speed else "N/A",
            "2nd Place": top_5[1]["rider_name"],
            "3rd Place": top_5[2]["rider_name"],
            "4th Place": top_5[3]["rider_name"],
            "5th Place": top_5[4]["rider_name"],
            "Time Gaps": ", ".join(time_gaps),
        })
        
    except Exception as e:
        print(f"Error fetching {race_name}: {e}")
        data.append({
            "Race": race_name,
            "Winner": "Error fetching data",
            "Team": "N/A",
            "Winner Time": "N/A",
            "Avg Speed (km/h)": "N/A",
            "2nd Place": "N/A",
            "3rd Place": "N/A",
            "4th Place": "N/A",
            "5th Place": "N/A",
            "Time Gaps": "N/A",
        })

df = pd.DataFrame(data)
df.to_excel("missing_Race_top5_finishers2.xlsx", index=False)
```
