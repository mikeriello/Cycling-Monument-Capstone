```python
from thefuzz import process
```

```python
def match_race_name(race_name, uci_proseries_cat):
    match, score = process.extractOne(race_name, uci_proseries_cat.keys())
    return uci_proseries_cat[match] if score > 85 else race_name
```

```python
core['Race Level'] = core['Race Name'].apply(lambda x: match_race_name(x, uci_proseries_cat))
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.procyclingstats.com/races.php?year=2024&month=&xnation=&filter=Filter&p=gravel&s=gravel-calendar"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

table = soup.find("table", {"class": "basic"})

headers = [th.text.strip() for th in table.find_all("th")]

data = []
for row in table.find_all("tr")[1:]:
    cells = row.find_all("td")
    data.append([cell.text.strip() for cell in cells])

df = pd.DataFrame(data, columns=headers)
df.to_excel("uci_gravel_race_classifications_2024.xlsx", index=False)
```
